```python
# connect to db
import mysql.connector

mydb = mysql.connector.connect(
  host="127.0.0.1",
  user=USER,
  password=PASSWORD,
  database="mpp21"
)

mycursor = mydb.cursor()
```

```python
newspapers = {}

with open('./newspapers-collected.csv', 'r') as f:
        reader = csv.reader(f)
        for r in reader:
            newspapers[r[0]] = r[1]
```

```python
import random

# db methods
def select_articles(curs, db, n):
    
    curs.execute("""SELECT article_id, title, metadata, content 
                FROM articles 
                WHERE newspaper = '""" + n + """' 
                AND is_privacy = 'privacy'
                """) 
    
    arts = {}
    result = curs.fetchall()
    for r in result:
        arts[r[0]] = {
            "title": r[1],
            #"summary": json.loads(r[2]), 
            "wordCount": json.loads(r[2])["wordCount"], 
            "content": r[3]
        }
    
    return arts

def get_annotation_sample(cnt, keys): # get 30 artoc;es
    
    copy = keys
    
    random_keys = []
    
    if len(keys) == 0:
        return []
    
    for i in range(cnt):
        choice = random.choice(copy)
        random_keys.append(choice)
        copy.remove(choice)
        
    return sorted(random_keys)
```

```python
articles = {}
annotations = {}

# for n in newspapers.keys():

articles["DT"] = select_articles(mycursor, mydb, "DT")
annotations["DT"] = get_annotation_sample(30, list(articles["DT"].keys()))
```

```python
import pandas as pd
df = pd.DataFrame(sorted(annotations["DT"]))
df.to_csv("htmls/dt_annotate_categories.csv", index=False, header=False)
```
