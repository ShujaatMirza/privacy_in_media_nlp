```python
import json
import pandas as pd
```

```python
from collections import defaultdict
```

```python
# dump verdicts to update all at once
n = "USA"

categories = ["collection", "processing", "dissemination"]
collection, processing, dissemination = {}, {}, {}
results = {
    "collection": collection,
    "processing": processing,
    "dissemination": dissemination
}


for c in categories:
    data = open("verdicts/"+ c + "_"+ n +".json", "r") 
    results[c] = json.load(data)
  
```

```python
results_final = {}

for a in results["processing"].keys():
    results_final[a] = {
        "collection": -1,
        "processing": -1,
        "dissemination": -1
    }
```

```python
for c in categories:
    
    for a in results[c].keys():
        results_final[a][c] = results[c][a][c]
```

```python
counts = {
        "collection": 0,
        "processing": 0,
        "dissemination": 0
    }

for a in results_final.keys():
    for c in categories:
        if results_final[a][c] == 1:
            counts[c] += 1
```

```python
pd.DataFrame.from_dict(counts, orient="index")
```
